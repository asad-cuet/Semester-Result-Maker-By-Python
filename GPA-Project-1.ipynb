```python
def inputFile(filename):
    import pandas as pd
    file=pd.read_excel("uploaded_files/"+filename)

    course_code=file.columns[1]
    

    course_title=file[course_code][0]

    course_credit=file[course_code][1]

    # Taking grade data file only
    df=pd.read_excel("uploaded_files/"+filename,skiprows=3,index_col=0)  # skip first 3 rows
    return {
            "course_code" : course_code,
            "course_title" : course_title,
            "course_credit": course_credit,
            "dataset":df
    }
```

```python
course_1 = inputFile(course_1)

course_2 = inputFile(course_2)

course_3 = inputFile(course_3)

course_4 = inputFile(course_4)

course_5 = inputFile(course_5)

course_6 = inputFile(course_6)

course_7 = inputFile(course_7)

course_8 = inputFile(course_8)
```

```python
result=[]
_dict={}
course_credit_list=[]
grade_point_list=[]
def getResult(course):
    _dict={}
    global result,student_id
    
    _dict[0]=course['course_code']
    _dict[1]=course['course_title']
    _dict[2]=course['course_credit']
    _dict[3]=course['dataset'].loc[student_id].loc['Grade']
    _dict[4]=course['dataset'].loc[student_id].loc['Grade Point']
    
    
    course_credit_list.append(course['course_credit'])
    grade_point_list.append(course['dataset'].loc[student_id].loc['Grade Point'])
    
    result.append(_dict)
```

```python
def getGPA(credit,grade):
    import numpy as np
    
    total_earned_credit=np.matmul(credit,grade)
    total_credit=np.sum(credit)
    
    gpa=total_earned_credit/total_credit
    gpa=round(gpa,3)
    
    return gpa
```

```python
getResult(course_1)
getResult(course_2)
getResult(course_3)
getResult(course_4)
getResult(course_5)
getResult(course_6)
getResult(course_7)
getResult(course_8)
```

```python
gpa=getGPA(course_credit_list,grade_point_list)
gpa
```
